7. Program to find Greatest Common Divisor of two numbers. For example, the GCDof20 and 28 is 4 and the GCD of 98 and 56 is 14.

```python
a=int(input("ENTER 1ST NUMBER"))
b=int(input("ENTER 2ND NUMBER"))
gcd = 1

for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        gcd = i

print("GCD is:", gcd)
```

8. Write a Python program to check whether a list contains a sublist

```python
main_list = [1, 2, 3, 4, 5, 6]
sub_list = [3, 4]

found = False

for i in range(len(main_list) - len(sub_list) + 1):
    if main_list[i:i+len(sub_list)] == sub_list:
        found = True
        break
if found:
    print("Sublist exists")
else:
    print("Sublist does not exist")
```

13. Sort a dictionary by value (ascending / descending)

```python
data = {'a': 3, 'b': 1, 'c': 2}

# Ascending
ascending = dict(sorted(data.items(), key=lambda x: x[1]))
print("Ascending:", ascending)

# Descending
descending = dict(sorted(data.items(), key=lambda x: x[1], reverse=True))
print("Descending:", descending)
```

15. Fibonacci Series up to n terms

```python
n = 7

a = 0
b = 1

fib_list = [a, b]

for i in range(2, n + 1):
    c = a + b
    fib_list.append(c)
    a = b
    b = c

print("First few Fibonacci numbers are:", *fib_list)
```

16. Count Frequencies in a List using Dictionary

```python
numbers = [1, 1, 1, 5, 5, 3, 1, 3, 3, 1, 4, 4, 4, 2, 2, 2, 2]

freq = {}

for num in numbers:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

for key in sorted(freq):
    print(key, ":", freq[key])
```

17. Sum of Odd and Even Series using Function

```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact


def sum_series(n):
    odd_sum = 0
    even_sum = 0

    # Odd series: 1²/1! + 3²/3! + 5²/5! ...
    for i in range(1, n + 1, 2):
        odd_sum += (i ** 2) / factorial(i)

    # Even series: 2²/2! + 4²/4! + 6²/6! ...
    for i in range(2, n + 1, 2):
        even_sum += (i ** 2) / factorial(i)

    print("Sum of Odd Series:", odd_sum)
    print("Sum of Even Series:", even_sum)


n = int(input("Enter value of n: "))
sum_series(n)
```

18. Factorial Using Recursion

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

num = int(input("Enter a number: "))
print("Factorial is:", factorial(num))
```

19. Function to Return Unique Elements from List

```python
def unique_list(lst):
    new_list = []
    for item in lst:
        if item not in new_list:
            new_list.append(item)
    return new_list


numbers = [1, 2, 2, 3, 4, 4, 5]
print("Unique elements:", unique_list(numbers))
```

20. Mini Project: Password Generator with User Class

```python
import random
import string


class User:
    users = []

    def __init__(self, user_id, name, password):
        self.details = (user_id, name, password)
        User.users.append(self.details)


def generate_password(words):
    try:
        if len(words) == 0:
            raise ValueError("Input cannot be empty!")

        base = random.choice(words)

        number = str(random.randint(10, 99))
        special = random.choice("!@#$%&*")
        capital = base.capitalize()

        password = capital + number + special

        if len(password) <= 8:
            password += random.choice(string.ascii_letters)

        return password

    except Exception as e:
        print("Error:", e)


# Taking input
try:
    user_id = int(input("Enter User ID: "))
    name = input("Enter Name: ")
    words = input("Enter some words separated by space: ").split()

    password = generate_password(words)

    if password:
        print("Generated Password:", password)
        user1 = User(user_id, name, password)
        print("User Details Stored:", user1.details)

except ValueError:
    print("Invalid input! User ID must be number.")
```
