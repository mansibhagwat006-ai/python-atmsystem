# **Comparison of Supervised and Unsupervised Learning**

| **Feature** | **Supervised Learning** | **Unsupervised Learning** |
|--------------|-------------------------|----------------------------|
| **Definition** | Learns from **labeled data**, where the correct output is already known. | Learns from **unlabeled data** and discovers hidden patterns or relationships. |
| **Training Data** | Requires **input data** along with the **correct target/output labels**. | Requires only **input data** without any target labels. |
| **Goal** | Predicts a class or numerical value (classification or regression). | Finds clusters, patterns, or associations in the data. |
| **Real-World App Example** | **Instagram Feed Ranking:** Predicts which posts a user is most likely to engage with based on previous interactions. | **Spotify Playlist Grouping:** Groups songs with similar characteristics (genre, mood, tempo) to create personalized playlists. |
| **Common Algorithms** | Linear Regression, Logistic Regression, Decision Tree, Random Forest, Support Vector Machine (SVM). | K-Means Clustering, Hierarchical Clustering, DBSCAN, Principal Component Analysis (PCA). |
| **Output** | Predicted labels or numerical values. | Clusters, groups, or hidden patterns. |
| **Example Applications** | Email spam detection, house price prediction, customer churn prediction, sentiment analysis. | Customer segmentation, recommendation systems, anomaly detection, document clustering. |

### **Conclusion**

- **Supervised Learning** uses **labeled data** to make predictions.
- **Unsupervised Learning** uses **unlabeled data** to discover hidden patterns or group similar data points.
