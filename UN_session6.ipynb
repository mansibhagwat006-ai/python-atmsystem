**1. Load the Iris dataset using Python's sklearn, perform hierarchical clustering 
with the 'ward' linkage, and plot the resulting dendrogram using scipy's dendrogram() function.**  

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
iris = load_iris()
X = iris.data
```

```python
Z = linkage(X, method='ward')
```

```python
plt.figure(figsize=(12,6))
dendrogram(Z)

plt.title("Hierarchical Clustering Dendrogram (Ward Linkage)")
plt.xlabel("Samples")
plt.ylabel("Euclidean Distance")
plt.show()
```

**2. Visually inspect the dendrogram you plotted and cut it at a height that results in
 3 distinct clusters. Assign each data point to its cluster and print the number of samples
 in each cluster.<br><br><em><strong>Hint:</strong> Use scipy's fcluster() with the 'criterion'
 set to 'distance' and your chosen height.</em>**

```python
from scipy.cluster.hierarchy import fcluster
import pandas as pd
```

```python
clusters = fcluster(Z, t=15, criterion='distance')
```

```python
df = pd.DataFrame(X, columns=iris.feature_names)
df['Cluster'] = clusters
```

```python
print(df['Cluster'].value_counts().sort_index())
```

**3. Repeat the dendrogram plotting for the same data using 'single' and 'complete'
 linkage methods, and compare how the cluster shapes and heights differ from 'ward' linkage.<br><br>
<em><strong>Hint:</strong> Place all three dendrograms side by side for visual comparison.</em>** 

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
methods = ['ward', 'single', 'complete']
```

```python
plt.figure(figsize=(18,5))

for i, method in enumerate(methods):
    plt.subplot(1,3,i+1)
    Z = linkage(X, method=method)
    dendrogram(Z)
    plt.title(f"{method.capitalize()} Linkage")
    plt.xlabel("Samples")
    plt.ylabel("Distance")

plt.tight_layout()
plt.show()
```
