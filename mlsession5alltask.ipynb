```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
instagram_hours = np.array([1.5, 2.0, 3.5, 4.0, 2.5, 5.0, 3.0])
battery_percentage = np.array([80, 75, 65, 60, 70, 50, 68])
```

```python
X = instagram_hours.reshape(-1, 1)
y = battery_percentage
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(instagram_hours, battery_percentage, s=100)
plt.title("Instagram Usage vs Phone Battery Percentage")
plt.xlabel("Instagram Usage (Hours per Day)")
plt.ylabel("Battery Percentage at End of Day (%)")
plt.grid(True)
plt.show()
```

```python
predicted_battery = model.predict(X)
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(instagram_hours, battery_percentage, color='blue', label='Actual Data')
plt.plot(instagram_hours, predicted_battery, color='red', linewidth=2, label='Regression Line')
plt.title('Instagram Usage vs Battery Percentage')
plt.xlabel('Instagram Usage (Hours per Day)')
plt.ylabel('Battery Percentage at End of Day (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
mse = mean_squared_error(y, predicted_battery)
```

```python
print("Mean Squared Error (MSE):", mse)
```
