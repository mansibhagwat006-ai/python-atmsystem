Design a Python class ATM that simulates the basic functionality of an Automated Teller Machine. Your solution should meet the following requirements: 
➔ Card Activation: Prompt the user for their full name and a valid 10-digit mobile number to activate the ATM card. 
➔ PIN Creation: Ask the user to create a 4-digit PIN and re-enter it for verification. 
➔ Initial Deposit: Ensure a minimum deposit of ₹1000 during activation.
➔ Main Menu: Provide options to: 
 ◆ Change PIN 
 ◆ Deposit Money 
 ◆ Withdraw Money (with applicable transaction charges) 
 ◆ Check Balance
 ◆ View Last 10 Transactions 
 ◆ Exit the ATM 
 ➔ Transaction Charges for withdrawals: 
 ◆ ₹0 for <= ₹1000 
 ◆ ₹100 for <= ₹20000 
 ◆ ₹1000 for <= ₹100000 
 ◆ ₹2000 for > ₹1000000 
 ➔ Error Handling: Handle invalid inputs and ensure secure PIN-based access for every operation. 
 ➔ Transaction History: Maintain a history of all deposits and withdrawals. Implement this functionality in a single Python class called ATM and demonstrate its usage. 

```python
class atm:
    def __init__(self):
        self.h=[]
        self.pin = 0
        self.valid = 0
        self.balance = 0
        self.name = ""
        self.no = 0
        self.card_activation()
        self.create_pin()

    def menu(self):
        user_input = input(
            """
        1. Press 1 to change the pin
        2. Press 2 to deposit money
        3. Press 3 to withdraw money
        4. Press 4 to check balance
        5. Press 5 to view last 10 transactions
        6. Press 6 to exit                        
                               """
        )
        if user_input == "1":
            self.change_pin()
        elif user_input == "2":
            self.deposit_money()
        elif user_input == "3":
            self.withdraw_money()
        elif user_input == "4":
            self.check_balance()
        elif user_input == "5":
            self.history()
        else:
            exit()

    def card_activation(self):
        name = input("Enter your full name for card activation:")
        self.name = name
        no = int(input("Enter your number:"))
        self.no = no
        a = len(str(no))
        if a == 10:
            print ("Your details are filled successfully ")
        else:
            print("Enter your details again")
            self.card_activation()
            
    def create_pin(self):  # 4 digit int number
        print("Hello, Welcome to ATM system")
        user_pin = int(input("Enter the 4 digit to set it as your pin"))
        self.user_pin = user_pin
        if len(str(self.user_pin))== 4:
            valid=int(input("Enter the 4 digit pin again"))
            self.valid = valid
            if valid == user_pin and len(str(valid))== 4:
                print("You pin has been created successfully")
                self.deposit_money()
        else :
            print("ENTER VALID PIN")
            self.create_pin()        

    def deposit_money(self):
        user_pin = int(input(
            "Enter the pin to deposit your money"))  # min 1000 rs deposit is necessary
        if user_pin == self.user_pin:
            amt = int(input("enter the amount that you want to deposit"))
            if self.balance == 0 and amt>1000:
                self.balance += amt
                print(f"Rs.{amt} has been deposited successfully")
                self.h.append(amt)
                self.menu()
            elif self.balance >= 1000:
                self.balance += amt
                print(f"Rs.{amt} has been deposited successfully")
                self.h.append(amt)
                self.menu()
            else:
                print("MINIMUM AMOUNT LIMIT IS 1000")
        else:
            print("YOU HAVE ENTERED THE WRONG PIN")
            self.deposit_money()

    def change_pin(self):
        user_pin = int(input("Enter the existing pin to change your pin"))
        if user_pin == self.user_pin:
            new_pin = input("Enter the new pin")
            self.user_pin = new_pin
            print("Pin has been changed successfully")
            self.menu()
        else:
            print("YOU HAVE ENTERED THE WRONG PIN")
            self.menu()

    def withdraw_money(self):  # min balance 1000
        user_pin = int(input("Enter the pin to withdraw the money"))
        if user_pin == self.user_pin:
            amt = int(input("Enter the amount you want to withdraw"))
            if self.balance > amt:
                if amt<=1000:
                    self.balance -= amt
                    print(f"Rs.{amt} has been withdrawed successfully")
                    self.h.append(amt)
                    self.menu()
                elif amt>1000 and amt<=20000 :
                    self.balance-=amt+100
                    print(f"Rs.{amt+100} has been withdrawed successfully which includes Transaction charge rs.100")
                    self.h.append(amt+100)
                    self.menu()
                elif amt>20000 and amt<=100000:
                    self.balance-=amt+1000
                    print(f"Rs.{amt+1000} has been withdrawed successfully which includes Transaction charge rs.1000")
                    self.h.append(amt+1000)
                    self.menu()
                elif self.balance> amt and amt>100000 and amt<=1000000:
                    self.balance-=amt+2000
                    print(f"Rs.{amt+2000} has been withdrawed successfully which includes Transaction charge rs.2000")
                    self.h.append(amt+2000)
                    self.menu()
                else:
                    print("You have insufficient balance")
                    self.menu()
        else:
            print("YOU HAVE ENTERED THE WRONG PIN")
            self.menu()

    def check_balance(self):
        user_pin = int(input("Enter your pin to check your balance"))
        if user_pin == self.user_pin:
            print(f"Available Balance: {self.balance}")
            self.menu()
        else:
            print("YOU HAVE ENTERED THE WRONG PIN")
            self.menu()
        
    def history(self):
        user_pin = int(input("Enter your pin to view history"))
        if user_pin == self.user_pin:
            print("The last 10 transaction history:",end="\n")
            for i in self.h[-10:]:
                print("--",i)
            self.menu()
        else:
            print("YOU HAVE ENTERED THE WRONG PIN")
            self.menu()
a = atm()
```
