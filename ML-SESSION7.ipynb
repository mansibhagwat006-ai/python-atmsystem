```python
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
```

```python
actual_times = [25, 30, 28, 35, 40, 22, 27, 33, 38, 29]
predicted_times = [26, 29, 30, 34, 39, 24, 28, 32, 37, 31]
```

```python
mse = mean_squared_error(actual_times, predicted_times)
rmse = np.sqrt(mse)
```

```python
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
actual_prices = [1200, 1500, 1800, 2200, 2500, 3000, 3500, 4000, 4500, 5000]
predicted_prices = [1250, 1450, 1750, 2300, 2450, 3100, 3400, 3900, 4600, 4900]
```

```python
def evaluate_model(actual, predicted):
    mae = mean_absolute_error(actual, predicted)
    r2 = r2_score(actual, predicted)
    print("Mean Absolute Error (MAE):", mae)
    print("R² Score:", r2)
evaluate_model(actual_prices, predicted_prices)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
y_test = np.array([78, 65, 89, 54, 72, 91, 83, 67, 75, 88])
pred_a = np.array([80, 63, 90, 52, 74, 89, 85, 66, 73, 90])
pred_b = np.array([75, 70, 82, 60, 68, 85, 79, 72, 70, 83])
```

```python
def evaluate_model(actual, predicted, model_name):
    mse = mean_squared_error(actual, predicted)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(actual, predicted)
    r2 = r2_score(actual, predicted)
```

```python
#print(f"\n{model_name}")
print("-" * 20)
print(f"MSE  : {mse:.2f}")
print(f"RMSE : {rmse:.2f}")
print(f"MAE  : {mae :.2f}")
print(f"R²   : {r2:.4f}")
```

```python
evaluate_model(y_test, pred_a, "Model A")
evaluate_model(y_test, pred_b, "Model B")
```

```python
train_rmse_a = 0.8
test_rmse_a = np.sqrt(mean_squared_error(y_test, pred_a))

train_rmse_b = 3.5
test_rmse_b = np.sqrt(mean_squared_error(y_test, pred_b))
```

```python
print("\nOverfitting Analysis")
print("-" * 20)

if abs(test_rmse_a - train_rmse_a) > abs(test_rmse_b - train_rmse_b):
    print("Model A is more likely overfitting.")
else:
    print("Model B is more likely overfitting")
```
